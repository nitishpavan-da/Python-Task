```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = np.random.randint(low = 0, high = 100, size = 100)
Y = np.random.randint(low = 0, high = 100, size = 100)
plt.scatter(X,Y,c = 'b',marker = "+")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Random Graph")
plt.show()
```

```python
A = np.arange(0, 2 * np.pi,0.1)
B = np.tan(A)
plt.plot(A,B,c='r')
plt.xlabel("Random Values")
plt.ylabel("Sine Values")
plt.title("Sine Wave")
plt.show()
```

```python
x = np.arange(0,2*np.pi,0.1)

plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
plt.title("Scatter Sine Wave")
y1 = np.sin(x)
plt.scatter(x,y1,c = 'r')

plt.subplot(2,2,2)
plt.title("Line Cosine Wave")
y2 = np.cos(x)
plt.plot(x,y2,c = 'b')

plt.subplot(2,2,3)
plt.title("Tangent Bar Graph")
y3 = np.tan(x)
plt.bar(x,y3)

plt.subplot(2,2,4)
plt.title("Box Plot Sine Wave")
y4 = np.sin(x)
plt.boxplot(y4)
```
