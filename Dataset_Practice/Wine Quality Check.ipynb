```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv("winenew.csv")
wine.head()
```

```python
wine.describe()
```

```python
wine.info()
```

```python
wine.isnull().sum()
```

```python
x = wine.iloc[:,:-1]
x
```

```python
y = wine.iloc[:,-1]
y.head()
```

```python
wine["quality"].value_counts()
```

```python
from sklearn import preprocessing
x = preprocessing.StandardScaler().fit_transform(x)
```

```python
x[0:4]
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnmodel = KNeighborsClassifier(n_neighbors = 3)
knnmodel.fit(x_train, y_train)
y_predict = knnmodel.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_predict, y_test)
acc
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_predict,y_test)
cm
```

```python
cm_tabular = pd.DataFrame(cm, columns = ["Good","Bad"])
cm_tabular
```

```python
predicted_y = pd.DataFrame(data = [y_test.values,y_predict])
predicted_y
```

```python
predicted_y.transpose()
```

```python
ks = 100
mean_acc = np.zeros((ks -1))

for i in range(1,ks):
    y_fit = KNeighborsClassifier(n_neighbors = i).fit(x_train, y_train)
    y_pred = y_fit.predict(x_test)
    mean_acc[i-1] = accuracy_score(y_pred,y_test)
```

```python
print(mean_acc)
```

```python
print("The max value of best mean values is ",mean_acc.max()," whose k value is ",mean_acc.argmax()+1)
```

```python
plt.plot(range(1,ks),mean_acc,'b')
plt.xlabel("Range of K values")
plt.ylabel("Mean Accuracy Values")
plt.title("Wine Quality Check")
plt.tight_layout()
```
