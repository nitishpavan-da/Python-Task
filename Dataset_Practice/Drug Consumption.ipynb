# Categorizing User and Non-User Drugs

```python
columns = ["Alcohol","Amphet","Amyl","Benzos","Caff","Cannabis","Choc","Coke","Crack","Ecstacy","Heroin","Ketamine","Legalh","LSD","Meth","Mushrooms","Nicotine","Semer","VSA"]
user_columns = ["user_Alcohol","user_Amphet","user_Amyl","user_Benzos","user_Caff","user_Cannabis","user_Choc","user_Coke","user_Crack","user_Ecstacy","user_Heroin","user_Ketamine","user_Legalh","user_LSD","user_Meth","user_Mushrooms","user_Nicotine","user_Semer","user_VSA"]
```

```python
for i in range(len(columns)):
    df_le.loc[((df_le[columns[i]] == 0) | (df_le[columns[i]] == 1)), user_columns[i]] = "Non User"
    df_le.loc[((df_le[columns[i]] == 2) | (df_le[columns[i]] == 3) | (df_le[columns[i]] == 4) | (df_le[columns[i]] == 5) | (df_le[columns[i]] == 6)), user_columns[i]] = "User"
```

```python
X1 = df_le.iloc[: , :13]
y1 = df_le.iloc[: , 13]
```

```python
y1.value_counts()
```

```python
from sklearn import preprocessing
X1 = preprocessing.StandardScaler().fit_transform(X1)
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.30, random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(X1_train, y1_train)
```

```python
user_nonuser_categories.classes_
```

```python
y_logregpred = logreg.predict(X1_test)
y_logregpred
```
